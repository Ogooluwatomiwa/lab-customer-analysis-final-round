```python
import pandas as pd
```

```python
data = pd.read_csv('marketing_customer_analysis.csv')
data
```

```python
data.describe()
```

```python
col = []
for column in data.columns:
    col.append(column.lower())
        
data.columns = col 
        
data.head()
```

```python
data.isna().sum()
round(data.isna()/len(data),4)*100
nulls_df = pd.DataFrame(round(data.isna().sum()/len(data),2)*100)

nulls_df = nulls_df.reset_index()
nulls_df
```

```python
data = data.dropna()
data = data.reset_index(drop=True)
data
```

```python
numerical = data.select_dtypes(include='number')
numerical
```

```python
categorical = data.select_dtypes(include=object)
categorical
```

```python
data.duplicated()
data.drop_duplicates()
```

```python
X = data.drop(['total claim amount'], axis=1)
y = data['total claim amount']
```

```python
X_num = X.select_dtypes(include='number')
X_num
```

```python
X_cat = X.select_dtypes(include=object)
X_cat
```

```python
from sklearn.preprocessing import MinMaxScaler
MinMaxtransformer = MinMaxScaler().fit(X_num)
X_normalized = MinMaxtransformer.transform(X_num)
print(X_normalized.shape)
X_normalized = pd.DataFrame(X_normalized,columns=X_num.columns)
X_normalized.head()
```

```python
X_cat = X_cat.drop(['customer', 'policy'], axis=1)
X_cat
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder().fit(X_cat)
print(encoder.categories_)
encoded = encoder.transform(X_cat).toarray()
print(encoded)
cols = [colname for row in encoder.categories_ for colname in row]
cols
onehot_encoded = pd.DataFrame(encoded, columns=cols)
cols_to_drop = [row[0] for row in encoder.categories_]
cols_to_drop
onehot_encoded = onehot_encoded.drop(cols_to_drop,axis=1)
X_cat_encoded = onehot_encoded
X_cat_encoded
```

```python
X_concat = pd.concat([X_normalized, X_cat_encoded], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_concat, y, test_size=0.2, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error as mae
from math import sqrt
```

```python
lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_train)

r2_train = r2_score(y_train, predictions)
print('train R2:',r2_train)
```

```python
predictions2 = lm.predict(X_test)

r2_test = r2_score(y_test, predictions2)
print('test R2:', r2_test)
```

```python
train_mse = mse(lm.predict(X_train), y_train)
print('train MSE:', train_mse)
```

```python
test_mse = mse(lm.predict(X_test), y_test)
print('test MSE:', test_mse)
```

```python
train_rmse = sqrt(train_mse)
print('train RMSE', train_rmse)
```

```python
test_rmse = sqrt(test_mse)
print('test RMSE', test_rmse)
```

```python
train_mae = mae(lm.predict(X_train), y_train)
print('train MAE', train_mae)
```

```python
test_mae = mae(lm.predict(X_test), y_test)
print('test MAE', test_mae)
```
